```python
df_best.Budget.fillna(0,inplace = True)
df_best.Votes.fillna(0,inplace = True)
```

```python
df_best.info()
```

```python
def best_worst(n,by,ascending = False, min_bud = 0, min_votes = 0):
    
    df2 = df_best.loc[(df_best.Budget >= min_bud) & (df_best.Votes >= min_votes), 
                      ['',by]].sort_values(by = by, ascending = ascending).head(n).copy()
    
    return HTML(df2.to_html(escape = False))
```

```python
#movies with highest revenues
best_worst(n = 5, by = 'Revenue')
```

```python
#movies with highest budget
best_worst(5,'Budget')
```

```python
#movies with highest profit
best_worst(5,'Profit')
```

```python
#movies with the worst profit
best_worst(5,'Profit', ascending = True)
```

```python
#Top 5 movies with the highest return on investment
best_worst(5,'ROI', min_bud = 5)
```

```python
#Top 5 movies with lowest ROI
best_worst(5,'ROI', ascending = True, min_bud = 20)
```

```python
#Movies with highest amount of votes
best_worst(5,'Votes')
```

```python
#highest rated movies
best_worst(5, 'Average Rating', min_votes = 50)
```

```python
#Movies with the lowest ratings
best_worst(5, 'Average Rating', ascending = True, min_votes = 100)
```

```python
#Lowest rated movies the biggest budgets
best_worst(5, 'Average Rating', ascending = True, min_votes = 100, min_bud = 50)
```
